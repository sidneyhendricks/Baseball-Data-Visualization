```python
park_state = parks1[['NAME', 'STATE']].rename(columns = {'NAME': 'park'}).set_index('park')
park_lon = parks1[['NAME', 'Longitude']].rename(columns = {'NAME': 'park'}).set_index('park')
park_lat = parks1[['NAME', 'Latitude']].rename(columns = {'NAME': 'park'}).set_index('park')
```

```python
state_dict = park_state['STATE'].to_dict()
lon_dict = park_lon['Longitude'].to_dict()
lat_dict = park_lat['Latitude'].to_dict()
```

```python
teams1['state'] = teams1['park'].map(state_dict)
teams1['longitude'] = teams1['park'].map(lon_dict)
teams1['latitude'] = teams1['park'].map(lat_dict)
```

```python
teams1.to_csv('../teams1.csv', index=False)
```
